## Exercice 2 Planification fiscale

```python
from pandas import DataFrame
import numpy as np
```

```python
df = DataFrame([
    {'product_id':23, 'name':'computer', 'wholesale_price':500, 'retail_price':1000, 'sales':100},
    {'product_id':96, 'name':'Python Cours', 'wholesale_price':35, 'retail_price':75, 'sales':1000},
    {'product_id':97, 'name':'Pandas Cours', 'wholesale_price':35, 'retail_price':75, 'sales':500},
    {'product_id':15, 'name':'banana', 'wholesale_price':0.5, 'retail_price':1, 'sales':200},
    {'product_id':87, 'name':'sandwich', 'wholesale_price':3, 'retail_price':5, 'sales':300}
])
df
```

```python
df['current_net']=(df['retail_price']- df['wholesale_price'])*df['sales']
```

```python
df
```

```python
df['after_15%']=df['current_net']*85
df
```

```python
df['after_20%']=df['current_net']*80
df['after_25%']=df['current_net']*75
df
```

```python
df[['current_net','after_25%']]
```

```python
df[['current_net','after_15%','after_20%','after_25%']].sum()
```

```python
df['current_net'].sum()- df[['after_15%','after_20%','after_25%']].sum()
```
