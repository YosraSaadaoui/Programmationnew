# Excercice 4 les quantiles

```python
import pandas as pd
import numpy as np
```

```python
df1=pd.read_csv(r'C:\Users\2340649\Documents\GitHub\Programmation2\dataFrame\data\taxi-distance.csv',
               header=0, names=['x']
              )
df1
```

```python
q1 = df1['x'].quantile(0.25)
q3 = df1['x'].quantile(0.75)
iqr = q3-q1 #Interquartile range
fence_low  = q1-1.5*iqr
fence_high = q3+1.5*iqr
df_out =df1.loc[(df1['x'] > fence_low) & (df1['x'] < fence_high)]
df_out
```
