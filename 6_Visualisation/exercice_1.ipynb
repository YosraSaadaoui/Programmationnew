```python
df=pd.read_json('C:/Users/thinkbook/Desktop/programmation2/cities.json')
df
```

```python
df.groupby('state')['city'].count().sort_values()
```

```python
df.groupby('state')['city'].count().sort_values().plot.bar(figsize=(10,10))
```
