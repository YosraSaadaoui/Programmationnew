```python
import pandas as pd
```

```python
filenames = ['C:/Users/thinkbook/Desktop/programmation2/nyc_taxi_2020-01.csv', 'C:/Users/thinkbook/Desktop/programmation2/nyc_taxi_2020-07.csv']
all_dfs = []
for one_filename in filenames:
    print(one_filename)
    all_dfs.append(pd.read_csv(one_filename, usecols = ['tpep_pickup_datetime', 'passenger_count', 'trip_distance', 'fare_amount','extra','mta_tax','tip_amount','tolls_amount', 'improvement_surcharge','total_amount','congestion_surcharge'], 
                               parse_dates = ['tpep_pickup_datetime']))    
```

```python
df=pd.concat(all_dfs)
df
```

```python
df['year']=df['tpep_pickup_datetime'].dt.year
df
```

```python
df['month']=df['tpep_pickup_datetime'].dt.month
df
```

```python
df['day']=df['tpep_pickup_datetime'].dt.day
```

```python
df=df[((df.month==1)|(df.month==7)) &(df.year==2020)]
df
```

```python
df=df.reset_index(drop=True)
df
```

```python
df=df.sample(frac=0.01)
```
